```python
#Euler problem 26
'''
Q:Find the value of d < 1000 for which 1/d contains the longest recurring cycle in its decimal fraction part.
A: create a function that is able to check recurring cycle for one number
then use a while loop to iterate the numbers from 2-999 in the function. 
The maths is that if the remainder recurs while doing the division, then there'll be a recurring number
So I needed to find  a way to know whether the remainder occurs during the division
i start from the very basics of division of numbers. If the numerator cannot be divided by the denominator,
we multiply the numerator by 10 and attempt the division while assigning a decimal to the quotient. We do this
till the numerator is now divisible by the denominator. Then we can assign the remainder to a list. 
As long as the remainder does not recur, we continue with the iteration till the remainder recurs, then 
we do not append that number but step out of the while loop
we initialized the list with None, so we can have indices for our values that start from 1. 
We can now deduct the index of the recurring number from the length of the list. 
This value will show us the length of the recurring decimals. 
'''

def repeat (d,num=1):
    remainder = []
    rem = None
    while rem not in remainder:
        remainder.append(rem)
        quot, rem = divmod((num*10), d)
        num = rem
    x = len(remainder) - remainder.index(rem)
    print(x)
    return x
    


i = 2
long = 0
f = 2
while i < 1000:
    x = repeat(i)
    if x > long:
        long = x
        f = i 
    i+=1
print(f'{f} has the longest recurring cycle of {long}')
```

```python
def spiral (num):
    gap_size = 1
    counter = 0
    four_counter = 4
    # a = []
    total = 0
    for i in range (1, num**2 + 1):
        if counter == 0:
            total += i
            # a.append(i)
            counter = gap_size
            four_counter -= 1
        elif counter != 0:
            counter -= 1
        if four_counter == 0:
            gap_size +=2
            four_counter = 4
    # print(a)
    # print(sum(a))
    print(total)

spiral (1001)

```

```python
def sum_spiral_diagonals(n):
    total_sum = 1  # Start with 1, which is in the center of the spiral
    for layer in range(1, (n // 2) + 1):
        square = (2 * layer + 1) ** 2
        step = 2 * layer
        total_sum += square
        total_sum += (square - step)
        total_sum += (square - 2 * step)
        total_sum += (square - 3 * step)
    return total_sum

# Example for a 1001x1001 spiral
result = sum_spiral_diagonals(1001)
print(result)
```
